### Creates a starting population of 100,000 numbers and calculates the mean of the population.

```python
np.random.seed(62)
population = np.random.normal(100, 40, 100000)
print("Population mean:", population.mean())
```

## Takes a random sample of 50,000 values from the population and checks mean.

### Mean of 50,000 (99.97) is 0.14 away from population mean (100.10)

```python
sample_50000 = np.random.choice(population, 50000)
print("Sample of 50,000 mean:", sample_50000.mean())
```

### Mean of 25,000 (99.78) is 0.33 away from population mean (100.10)

```python
sample_25000 = np.random.choice(population, 25000)
print("Sample of 25,000 mean:",sample_25000.mean())
```

### Mean of 12,500 (99.86) is 0.24 away from population mean (100.10)

```python
sample_12500 = np.random.choice(population, 12500)
print("Sample of 12,500 mean:", sample_12500.mean())
```

### Mean of 6,250 (100.35) is 0.24 away from population mean (100.10)

```python
sample_6250 = np.random.choice(population, 6250)
print("Sample of 6,250 mean:", sample_6250.mean())
```

### Mean of 3,125 (99.38) is 0.73 away from population mean (100.10)

```python
sample_3125 = np.random.choice(population, 3125)
print("Sample of 3,125 mean:", sample_3125.mean())
```

### Mean of 1,000 (99.21) is 0.90 away from population mean (100.10)

```python
sample_1000 = np.random.choice(population, 1000)
print("Sample of 1,000 mean:", sample_1000.mean())
```

### Mean of 500 (97.10) is 3.01 away from population mean (100.10)

```python
sample_500 = np.random.choice(population, 500)
print("Sample of 500 mean:", sample_500.mean())
```

## Summary -- As the sample size decreases from 50,000 down to 500, the sample mean becomes less reliable and gets further away from the original true population mean of 100.10.

# Demonstrating the difference between mean, median, mode, variance, standard deviation, uniform distribution, and normal distribution.

## Imports packages, then sends a request to the OpenIntro website to download the oscars.csv file

```python
import pandas as pd
import plotly.express as px
import plotly.io as pio
pio.renderers.default = "iframe"
import requests
from io import StringIO

url = "https://www.openintro.org/data/csv/oscars.csv"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
response = requests.get(url, headers=headers)
response.raise_for_status()
df = pd.read_csv(StringIO(response.text))
```

## df is the dataframe with the Oscar awards data taken from openinto data sets.

#### This dataset contains variables - winner's name, age, birthplace, birth date, and the movie they won for.
#### We are going to use the "age" column to demonstrate mean, median, mode, variance, standard deviation, and compare it against simulated uniform and normal distributions

```python
df
```

### Mean - the mean is the average value in the data

#### Uses pandas .mean() function to calculate the mean age from the dataset.

```python
mean_age = df["age"].mean()
print("The mean age of an Oscar Acting Winner is", round(mean_age, 2))
```

### Mean - the median is the middle value for a data set in order

#### Uses pandas .median() function to calculate the median age from the dataset.

```python
median_age = df["age"].median()
print("The median age of an Oscar Acting Winner is", median_age)
```

### Mode - the mode is value in the data set that appears the most frequently

#### Uses pandas .mode() function to get the most frequent age values from the dataset

```python
mode_age = df["age"].mode()[0]
mode_age
```

### Variance - the variance is the measure of the spread of a data set relative to its mean

#### Uses pandas .var() function to get the variance of the "age" column

```python
variance = df["age"].var()
variance
```

### Standard Deviation - the standard deviation is the average distance between data values and the mean

#### Uses pandas .std() function to calculate the standard deviation of age from the dataset.

```python
standard_dev = df["age"].std()
standard_dev
```

### Uniform Distribution - distribution where the data is spread roughly symmetric and constant

#### Creates a histogram with a random distribution with 1000 random values to simulate a uniform distribution.

```python
uniform_distribution = np.random.uniform(low=df["age"].min(), high=df["age"].max(), size=1000)

fig = px.histogram(uniform_distribution, nbins=20, title="Uniform Distribution")
fig.show()
```

### Normal Distribution - distribution where the data is symmetric and centered around the mean.

#### Creates a histogram with a random distribution with 1000 random values to simulate a normal distribution.

```python
normal_distribution = np.random.normal(loc=mean_age, scale=standard_dev, size=1000)

fig = px.histogram(normal_distribution, nbins=20, title="Normal Distribution")
fig.show()
```
