## What It's About
The paper attempts to predict the count of Olympic medals each country will win based on various factors such as: country GDP, number of athletes, and competition level. The goal of the research is to help countries better allocate sports resources to prepare for the 2028 LA Olympics.

## Linear Regression Used
- X variables (independent variables): GDP, athlete competition level, participation numbers, gender ratio, host country status
- Y variable (dependent variable): Total number of gold or other Olympic medals won
- They use multiple linear regression with beta representing multiple linear slopes.
